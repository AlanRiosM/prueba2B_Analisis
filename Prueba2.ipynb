```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("dfCaso1.csv")
```

```python
director_movie = df['Year'].value_counts()
plt.pie(director_movie, labels=director_movie.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Porcentajes de peliculas por aos')
plt.show()
```

```python
director_movie = df['Director'].value_counts().head(25)
director_movie.plot(kind='bar', color='skyblue')
plt.title('Directores con el numero de pelicula')
plt.xlabel('Directores')
plt.ylabel('Nmero de peliculas')
plt.show()
```

```python
df2 = df.head(20)
movie_metascore = df2.groupby('Title')['Metascore'].sum()
movie_metascore.plot(kind='bar', color='yellow')
plt.title('Peliculas - Metascore')
plt.xlabel('Peliculas')
plt.ylabel('Metascore')
plt.show()
```

```python
year_runtime = df.groupby('Year')['Runtime (Minutes)'].sum()
year_runtime.plot(kind='bar', color='blue')
plt.title('Aos - Minutos')
plt.xlabel('Aos')
plt.ylabel('Minutos')
plt.show()
```

```python
title_revenue = df.groupby('Title')['Revenue (Millions)'].sum().head(25)
title_revenue.plot(kind='bar', color='red')
plt.title('Peliculas - Ganancias(Millones)')
plt.xlabel('Ganancias(Millones)')
plt.ylabel('Peliculas')
plt.show()
```

```python
genre_movie = df['Genre'].value_counts().head(10)
plt.pie(genre_movie, labels=genre_movie.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Porcentajes de peliculas por aos')
plt.show()
```

```python
director_runtime = df.groupby('Director')['Runtime (Minutes)'].sum().head(25)
director_runtime.plot(kind='bar', color='brown')
plt.title('Director - Minutos')
plt.xlabel('Director')
plt.ylabel('Minutos')
plt.show()
```

```python
director_votes = df.groupby('Director')['Votes'].sum().head(25)
director_votes.plot(kind='bar', color='gray')
plt.title('Director - Votos')
plt.xlabel('Director')
plt.ylabel('Votos')
plt.show()
```

```python
director_genre = df.groupby('Genre')['Director'].value_counts().head(25)
director_genre.plot(kind='bar', color='pink')
plt.title('Genero - Numero de Peliculas')
plt.xlabel('Genero')
plt.ylabel('Numero de Peliculas')
plt.show()
```

```python
year_rating = df.groupby('Year')['Rating'].mean().head(25)
year_rating.plot(kind='bar', color='pink')
plt.title('Aos - Rating')
plt.xlabel('Aos')
plt.ylabel('Rating')
plt.show()
```
